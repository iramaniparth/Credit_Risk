```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as mlt
import seaborn as sns
mlt.style.use('fivethirtyeight')
from scipy import stats
from scipy.stats import norm,skew
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('amex.csv')
data.drop(['Unnamed: 0','mvar16','mvar19','mvar17'],axis=1,inplace=True)
data.head()
X_sub=pd.read_csv('amextest.csv')
X_sub.drop(['Unnamed: 0','mvar16','mvar19','mvar17'],axis=1,inplace=True)
data['mvar47'].loc[data['mvar47'] == 'L'] = 0
data['mvar47'].loc[data['mvar47'] == 'C'] = 1
X_sub['mvar47'].loc[X_sub['mvar47'] == 'L'] = 0
X_sub['mvar47'].loc[X_sub['mvar47'] == 'C'] = 1
kall = data.iloc[:,0:49]
kalltr = kall.dropna()

values = kall.values

from sklearn.preprocessing import Imputer
imputer = Imputer()

transformed_values = imputer.fit_transform(values)

X = pd.DataFrame(transformed_values)
X.columns = kall.columns

for index in X.columns:
    data[index]=X[index]
```

```python
data.iloc[:,:44]
```

```python
data.head()
data0=np.array(data)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data0[:,0:45], data0[:,45], random_state=0)
print(data0.shape)
data0[:,45]
```

```python
import lightgbm as lgb
```

```python
train_data=lgb.Dataset(X_train,label=y_train)
```

```python
#setting parameters for lightgbm
param = {'num_leaves':300, 'objective':'binary','max_depth':5,'learning_rate':.05,'max_bin':200}
param['metric'] = ['auc', 'binary_logloss']
```

```python
from datetime import datetime
#training our model using light gbm
num_round=300
start=datetime.now()
lgbm=lgb.train(param,train_data,num_round)
stop=datetime.now()
```

```python
execution_time_lgbm = stop-start
execution_time_lgbm
```

```python
ypred2=lgbm.predict(X_test)
```

```python
for i in range(ypred2.shape[0]): 
    if ypred2[i]>=.5:       # setting threshold to .5 
       ypred2[i]=1 
    else: 
       ypred2[i]=0  
```

```python
from sklearn.metrics import accuracy_score 
accuracy_lgbm = accuracy_score(ypred2,y_test)
accuracy_lgbm
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbrt1 = GradientBoostingClassifier(random_state=0,max_depth=4, learning_rate=0.1)
gbrt1.fit(X_train, y_train)
print("Accuracy on training set: {:.6f}".format(gbrt1.score(X_train, y_train)))
print("Accuracy on test set: {:.6f}".format(gbrt1.score(X_test, y_test)))
```
